```
#7(ii)
def prime(n):
  if n<=1:
    return False
  for i in range(2,int(n**0.5)+1):
    if n%i==0:
      return False
  return True

prime_numbers=[x for x in range(600,801)if prime(x)]

print("The prime numbers between 600 and 800 are:",prime_numbers)
```

```
#10th
def is_even(number):
    return number % 2 == 0

def main():
    try:
        number = int(input("Input a number: "))
        if is_even(number):
            print(f"{number} is an even number.")
        else:
            print(f"{number} is an odd number.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

if __name__ == "__main__":
    main()
```
